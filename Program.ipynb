```python
product=[]
a=[]
b={}
goods=[]
newGoods=[]
n=[]
```

```python
def addProduct():#1.입고 물품 추가 함수
    print('입고 물품을 입력하세요.')
    my_product={}
    my_product['품명']=str(input('품명 : '))
    my_product['수량']=int(input('수량 : '))
    my_product['단가']=int(input('단가 : '))
    my_product['유통기한(월)']=int(input('유통기한(월) : '))
    my_product['유통기한(일)']=int(input('유통기한(일) : '))
    print('(알림) 물품 ',my_product['품명'],'(이)가 정상 입력 되었습니다.')
    product.append(my_product)
```

```python
def printProduct():#2.물품 리스트 출력 함수
    if not product:
        print('(알림) 저장된 물품이 없습니다.')
    else:
        print('=============물품 리스트==========')
        print('번호    품명    수량      유통기한')
        for index,item in enumerate(product):
            print(index+1,'    ',product[index]['품명'],'',product[index]['수량'],'   ',product[index]['유통기한(월)'],'월',product[index]['유통기한(일)'],'일')
```

```python
def trashProduct():#3.물품 폐기 함수
    print('오늘의 날짜를 입력하세요')
    month=int(input('월 : '))
    day=int(input('일 : '))
    print('============폐기 물품 리스트========')
    print('번호    품명      수량      유통기한')
    num=1
    for index,item in enumerate(product):
        if product[index]['유통기한(월)']<month:
            print(num,'    ',product[index]['품명'],'',product[index]['수량'],'   ',product[index]['유통기한(월)'],'월',product[index]['유통기한(일)'],'일')
            num+=1
            del product[index]
        elif product[index]['유통기한(월)']==month:
            if product[index]['유통기한(일)']<=day:
                print(num,'    ',product[index]['품명'],'',product[index]['수량'],'   ',product[index]['유통기한(월)'],'월',product[index]['유통기한(일)'],'일')
                num+=1
                del product[index]
            
    print('(알림)',month,'월',day,'일자로 정상 폐기 되었습니다.')
```

```python
def billProduct():#4. 손님 계산 함수
    for i in range(len(product)):
        a.append(product[i]['품명'])
    for k in range(len(product)):
        b[product[k]['품명']]=product[k]['수량']
    for r in range(len(product)):
        b[product[r]['수량']]=product[r]['단가']
    print('손님의 구매 희망 물품 및 수량을 입력하세요.')
    while True:
        clients_goods={}
        clients_goods['품명']=str(input('품명 : '))
        goods.append(clients_goods)
        if clients_goods['품명'] == '구매끝':
            break
    
        elif clients_goods['품명'] not in a:
            print('(알림) 해당물품 재고 없음')
            goods.remove({'품명': clients_goods['품명']})
            continue
        clients_goods['수량']=int(input('수량 : '))
        goods.append(clients_goods)
        if clients_goods['수량'] > int(b[clients_goods['품명']]):
            print('(알림) 해당물품 수량 부족')
            for i in range(2):
                goods.remove({'품명': clients_goods['품명'], '수량': clients_goods['수량']})
    for x in goods:
        if x not in newGoods:
            newGoods.append(x)
    for e in range(len(product)):
        if newGoods[e]['품명'] == product[e]['품명']:
            newGoods[e]['단가'] = b[product[e]['수량']]
    for g in range(len(product)):
        n.append(newGoods[g]['단가']*newGoods[g]['수량'])
    print('=======================================')
    print('손님의 구매 희망 물품은 다음과 같습니다.')
    for p in range(len(product)): 
        print(newGoods[p]['품명'],'',newGoods[p]['수량'],'개')
    print('총 결제 금액은','',sum(n),'원 입니다.')
    print('=======================================')
    question=int(input('통신사 할인을 받으시겠습니까?(예 :1/아니오 :2) : '))
    if question==1:
        salePay=int(sum(n)*(85/100))
        print('(통신사 할인 15% 적용) 구매 희망 물품의 결제 금액은',salePay,'원 입니다.')
        client_pay=int(input('고객 지불 금액 입력 : '))
        if client_pay<salePay:
            re_client_pay=int(input('(알림) 금액이 부족합니다. 다시 입력하세요. : '))
            if re_client_pay>=salePay:
                print('거스름돈은',re_client_pay-salePay,'원 입니다. 정상구입 되었습니다. 감사합니다.')
        else:
            print('거스름돈은',client_pay-salePay,'원 입니다. 정상구입 되었습니다. 감사합니다.')
        
    if question==2:
        client_pay=int(input('고객 지불 금액 입력 : '))
        if client_pay<sum(n):
            re_client_pay=int(input('(알림) 금액이 부족합니다. 다시 입력하세요. : '))
            if re_client_pay>=sum(n):
                print('거스름돈은',re_client_pay-sum(n),'원 입니다. 정상구입 되었습니다. 감사합니다.')
        else:
            print('거스름돈은',client_pay-sum(n),'원 입니다. 정상구입 되었습니다. 감사합니다.')
            
    for t in range(len(product)):
        if newGoods[t]['품명'] == product[t]['품명']:
            product[t]['수량'] = product[t]['수량']-newGoods[t]['수량']
```

```python
while True:
    print('===============================')
    selectNum=int(input('동작을 선택하세요 (1.입고 물품 추가 / 2. 물품 리스트 출력 / 3. 물품 폐기 / 4. 손님 계산 / 5. 종료):'))
    if(selectNum==1):
        addProduct()
    elif(selectNum==2):
        printProduct()
    elif(selectNum==3):
        trashProduct()
    elif(selectNum==4):
        billProduct()
    else:
        print('프로그램을 종료합니다.')
        break
```
